# 튜플(tuple)
* tuple: 저장된 원소들을 변경할 수 없는 리스트.
* 인덱스 기반의 데이터타입.
 * 인덱스를 사용해서 저장된 원소를 참조.
 * indexing, slicing
* 원소를 변경할 수 없기 때문에, append,remove 등과 같은 메소드가 제공되지 않음.

```python
numbers = (1,2,3,1,2,3,1,2)
print(numbers)
print(type(numbers))
```

```python
numbers.count(1) #tuple numbers에서 값 1이 몇 번 등장하는지를 리턴.
```

```python
numbers.index(1) #tuple numbers에서 값1이 가장 처음 등장하는 인덱스를 리턴.
```

# indexing

```python
print(numbers)
```

```python
numbers[0] #여기에서 []는 리스트를 뜻하는게 아닌 인덱스를 뜻하는 것.
```

# slicing

```python
numbers[:3] # tuple에서 앞에서 3개를 slicing
```

```python
numbers[-3:]
```

## decomposition(분해)

```python
a, b = (1,2) # 튜플의 원소 개수만큼 변수를 선언.
print(a)
print(b)
```

```python
x, y, z =[1,2,3]
print(x,y,z)
```

## tuple과 for 반복문

```python
for x in (1,2,10,20):
    print(x)
```

```python
mat_tuple = ((1,2),
             (3,4),
             (5,6))
for row in mat_tuple:
    for x in row:
        print(x, end='\t')
    print()
```

```python
for x, y in mat_tuple: #각각의 원소의 갯수가 같을 때
    print(x,y, sep='\t')
```
