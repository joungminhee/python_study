# Python Data Type (자료타입)



*   숫자 타입
    * 정수 (int)
    * 실수 (float)
*   문자열(str)
*   논리 타입(bool): True 또는 False 두 가지 상태를 저장하는 변수.



## 숫자 타입

````python
n = 100 # 변수 = 값
print(type(n)) # 변수 n의 자료 타입을 출력
print('n = ', n)
````

````python
n = 3.14
print(type(n))
print('n = ' , n)
````

## 논리타입

````python
b = True # 파이썬의 예약어(keyword)는 대소문자를 구분함!
print(type(b))
print(b)
````

````python
a = False # 파이썬의 예약어(keyword)는 대소문자를 구분함!
print(type(a))
print(a)
````

````python
x = 5
result = x>0
print(result)
````

## 명시적 타입 변환 (explicit type conversion/casting)


*   `int(값)` : 값을 정수로 변환해서 리턴하는 함수.
*   `float(값)` : 값을 실수로 변환해서 리턴하는 함수.
*   `str(값)` : 값을 문자열로 변환해서 리턴하는 함수.




````python
n_str1 = '123'
n_str2 = '100'
print(type(n_str1)) # 변수 n_str1 에 저장된 값의 자료타입은 'str(문자열)'
print(n_str1+n_str2) # 문자열 + 문자열 => 문자열 이어붙이기(concatenate)
````

````python
n1 = int(n_str1)
n2 = int(n_str2)
print(type(n1))
print(n1+n2)  # 정수 + 정수 => 숫자들의 덧셈
````

````python
pi = '3.14'
print(type(pi))
pi = float(pi) # 변수 pi에 저장된 값을 실수(pi)타입으로 변환해서 다시 변수 pi에 저장
print(type(pi))
````

````python
number = input('숫자를 입력하세요>>>')
number = float(number)
print(type(number))
````

````python
number = float(input('숫자를 입력하세요>>>'))
print(type(number))
````

# 연산자(Operator)

## 산술연산자 (arithmetic operators)

* 종류 : `+,-,*,/,**,//,%` #// 몫만 계산해라 , % 나머지만 계산해라
* 일반적으로 숫자 타입(int, float)의 값에서 사용 가능.


````python
x,y = 10,3

print(f'{x}+{y}={x + y}')
print(f'{x}-{y}={x - y}')
print(f'{x}*{y}={x * y}')
print(f'{x}/{y}={x / y}')
print(f'{x}의{y} 거듭제곱 = {x ** y}')
print(f'{x}의{y}로 나눈 몫 = {x // y}')
print(f'{x}의{y}로 나눈 나머지 = {x % y}')
````

문자열 (str) 타입에서 사용할 수 있는 산술 연산자 :
* `+` : concatenate. 문자열 이어 붙이기.
* `*` : replicate. 문자열을 곱한 숫자만큼 복제하기.

````python
print ('Hello' + 'python')
````

````python
print ('Hello' * 5)
````

bool 타입의 값들을 산술 연산을 할 때는 True는 1, False는 0으로 자동 변환된 후 
산술 연산이 수행됨.

````python
print(True + True)
````

````python
(True+True+True+False+True)/5
````

## 비교 연산자와 논리 연산자
* 비교 연산자 : `==,!=,>,>=,<,<=`, ...
* 논리 연산자 : `and, or, not `


## 할당 연산자, 복합 할당 연산자
* `=`: 할당 연산자. 왼쪽의 값을 오른쪽의 변수에 저장
* 복합 할당 연산자 : 산술 연산자와 할당 연산자가 합쳐서 만들어진 연산자.
 * `+=, -=, *=,/=,//=,%=,...` 

````python
x = 1 #정수값 1을 변수 x에 저장(할당)
x += 1 # (1) x+1, (2) 결과를 변수 x에 저장(할당)
print(x)
````
