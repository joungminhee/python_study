# List(리스트)

* list(리스트) : 여러 개의 값들을 하나의 변수에 저장하기 위한 데이터 타입.
* elememt(원소, 요소): 리스트에 저장된 각각의 값
* index(인덱스) : 리스트에서 각각의 원소들이 저장된 위치.
 * 리스트의 인덱스는 **0**부터 시작
 * 리스트의 마지막 원소의 인덱스는 (원소 개수 -1)
* 리스트는 원소들을 변경(추가, 삭제)할 수 있음
 * 튜플(tuple)과 다른 점!

````python
# 숫자들을 저장하는 list
numbers = [1,2,10,20,100]
print(numbers)
````

````python
print(type(numbers))
````

## indexing

인덱스를 사용해서 리스트의 원소를 사용하는 방법

````python
print(numbers[0]) #리스트에서 첫번째 원소
````

````python
print(numbers[4]) #리스트에서 마지막 원소
````

````python
print(numbers[-1])
````

## slicing

```
list_name[start:end]
```

* 리스트에서 start <= index < end 범위의 원소들을 선택
* start를 생략하면 인덱스 0번 원소(리스트의 첫번째 원소)부터 선택.
* end를 생략하면 리스트의 마지막 원소까지 선택.
* slicing의 결과는 **새로운 리스트**!!
 * 리스트를 자르면 리스트가 된다.

````python
print(numbers)
````

````python
print(numbers[1:4]) #range랑 동일 1<= index < 4
````

````python
print(numbers[2:]) #end를 생략하면 리스트의 끝까지 자른다
````

````python
print(numbers[:2]) # start를 생략하면 리스트의 처음부터 end인덱스까지 slicing
````

````python
# 리스트에서 끝에서 2개 원소 선택
print(numbers[-2:])
````

## 문자열

문자들의 리스트.
* 인덱스를 사용할 수 있음.
* indexing: 인덱스를 사용해서 특정 위치에 있는 문자 한 개를 선택할 수 있음.
* slicing: 인덱스를 사용해서 문자열을 잘라낼 수 있음.

````python
message = '안녕하세요, Python!' #14개의 문자를 가지고 있는 문자열.
print(len(message)) #문자열의 길이
````

````python
# message 문자열에서 첫번째 문자 한 개를 출력
print(message[0])
````

````python
# 문자열에서 첫 5개 문자들을 선택
print(message[:5])
````

````python
# 문자열에서 마지막 5개 문자들을 선택
print(message[-5:])
````

## list의 기능들 - 함수 (메서드, method)

### 원소(element) 추가

````python
numbers = [1,2,3,4,5]
numbers.append(100) # list.append(값) : 값을 list의 가장 마지막에 추가.
print(numbers)
````

````python
# list.insert(index,object): list의 index위치에 object(값)을 추가
````

````python
numbers.insert(0,123)
print(numbers)
````

````python
# list.extend(iterable): 리스트 이어붙이기(conacatenate) '+'연산과 동일
numbers1 = [1,2,3]
numbers2 = [10,20,30]
numbers1.extend(numbers2) # numbers1+numbers2
print(numbers1)
````

````python
numbers1 = [1,2,3]
numbers2 = [10,20,30]
numbers1.append(numbers2) #append 메서드는 리스트를 '한 개'의 원소로 추가함!
print(len(numbers1))
print(numbers1)
````

### 원소삭제

````python
# list.remove(값): 리스트에서 처음 등장하는 값 한 개를 삭제.
numbers = [1,2,3,1,2,10,20]
numbers.remove(2)
print(numbers)
````

````python
# list.pop(index) : index 위치의 원소를 삭제. index를 전달하지 않으면 가장 마지막 원소를 삭제.
numbers = [1,2,10,20,100,200]
numbers.pop(0) #리스트의 첫번째 원소(인덱스 0번 원소)를 삭제
print(numbers)
````

````python
numbers.pop() #index를 전달하지 않으면 가장 마지막 원소를 삭제.
print(numbers)
````

````python
# list.clear() : 리스트의 모든 원소를 삭제
numbers.clear()
print(numbers)
````

````python
#원소대체
numbers=[1,2,3]
numbers[1] = 100
print(numbers)
````
