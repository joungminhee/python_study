# 난수 (random number)만드는 방법

```python
import random # 난수를 만들어 주는 함수들을 가지고 있는 모듈(파이썬 함수가 정의 되어있는 파일)
```

```python
# random.random() : 0 <= r <1 범위의 (실수) 난수들을 생성.
random.random() *10
```

```python
# random.randrange(start, stop): start <= x < stop 범위의 정수로 만들어진 난수들을 생성.
random.randrange(1,4)
```

```python
# random.randrange(stop): 0 <= x < stop 범위의 정수 난수들을 생성.
random.randrange(4)
```

# 가위,바위,보 게임 승패 결정

* 가위 (1), 바위(2), 보(3)
* 컴퓨터 - 1,2,3 중의 숫자 하나를 난수로 저장
* 사용자 - 1,2,3 중의 숫자 하나를 입력받아서 저장
* 가위/바위/보 게임 결과를 조건문을 사용해서 출력

```python
computer = random.randrange(1,4)
print(computer)
```

```python
print('===가위/바위/보===')
print('[1] 가위')
print('[2] 바위')
print('[3] 보')
user = int(input('선택>>>'))
```

```python
# 게임 결과 출력
if computer == user:
    print('draw')
else:
    if user == 1:
        if computer == 2:
            print ('컴퓨터win')
        else:
            print ('유저win')
    if user == 2:
        if computer == 3:
            print ('컴퓨터win')
        else:
            print ('유저win')
    if user == 3:
        if computer == 1:
            print ('컴퓨터win')
        else:
            print ('유저win')
```

```python
#비김
#elif
#else


if user == computer: #비김
    print('draw')
elif (user ==1 and computer ==3) or (user ==2 and computer ==1) or (user ==3 and computer ==2): #user 이김 (elif)
    print('user win')
else: #computer 이김
    print('computer win')
```
