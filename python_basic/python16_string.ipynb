# str 클래스의 메서드

```python
s = 'heLLo pyTHon!' # 문자열 클래스의 인스턴스 생성
```

```python
print(type(s))
```

```python
result = s.lower() # 소문자로 변환된 "새로운 문자열" 을 리턴함.
print(result)
print(s) # 원본 문자열은 변경되지 않음.
```

```python
s.upper() # 대문자로 변환된 "새로운 문자열" 을 리턴함.
```

```python
s.lower().upper() #메서드 연쇄 호출
```

```python
s.count('p') #찾을때도 대/소문자 구별
# 원본 문자열 s에서 문자열의 일부 sub가 몇 번 등장하는지를 리턴.
```

```python
url = 'https://www.google.com'
url.startswith('http') # 문자열이 패턴으로 시작하면 True, 그렇지 않으면 False를 리턴.
```

```python
url.endswith('co.kr')
```

```python
#str.find(패턴): 문자열에서 패턴을 찾을 수 있는 가장 낮은 (작은) 인덱스를 리턴.
s = '안녕하세요, 파이썬 참 쉽죠? 파이썬 열심히 공부합시다!'
s.find('파이썬')
```

```python
email = 'jake@gmail.com'
# email 주소에서 이메일 아이디와 이메일 도메인 주소

idx = email.find('@')
print(idx)

print(email[0:idx])
print(email[idx+1:])
```

```python
#str.split(): 문자열을 공백(whitespace: space, tab, enter)
#을 기준으로 문자열을 잘라서 만들어진 단어들의 리스트를 리턴.
print(s)
print(s.split())
```

```python
date = '2022-05-18'
# 문자열 date에서 [년,월,일]을 갖는 리스트를 만듦.
date.split('-')
```

```python
csv = '홍길동, 100,90,95'
# csv의 내용을 쉼표(,)를 기준으로 자른 리스트
csv.split(',')
```

```python
# str.join(iterable): 문자열 str을 사용해서 iterable 안의 문자열을 이어붙임.
'.'.join (['ab','pq','rs'])
```

```python
'/'.join (['2022', '05','18'])
```
