# list와 for 반복문

```python
numbers = [1,10,100,1000]
# list의 값들을 순회(반복, iteration)
for x in numbers:
    print(x)
```

```python
# list의 index와 값들을 함께 순회(iteration)
for index, value in enumerate(numbers): #enumerate = index와 value 동시에 꺼내주는 역할
    print(index, ':', value)
```

* 빈 리스트(empty list)를 만듦.
* 리스트에 정수인 난수 10개를 저장.
* 리스트에서 짝수들로만 이루어진 새로운 리스트를 만듦.
* 리스트에서 홀수들로만 이루어진 새로운 리스트를 만듦.

```python
import random #난수를 사용하기 위한 모듈
```

```python
numbers = []
for _ in range(10):
    # 0 이상 100 미만의 정수 난수를 리스트 numbers의 가장 마지막에 추가.
    numbers.append(random.randrange(100)) # randrange 함수는 지정된 구간에 포함되어있는 숫자들 중에서 임의의 1개를 무작위로 뽑아서 돌려주는 난수발생 함수
print(numbers)
```

```python
evens = [] # 짝수들만 저장할 리스트
odds = [] #홀수들만 저장할 리스트
for x in numbers:
    if x % 2 ==0 : #리스트에서 꺼낸 원소가 짝수이면
        evens.append(x)
    else: #꺼낸 원소가 홀수이면
        odds.append(x)
print(evens)
print(odds)
```

```python
integers=[]
for x in range(10):
    integers.append(random.randrange(-10,10))
print(integers)

positives=[]
for y in integers:
    if y > 0:
        positives.append(True) # 'True' 'ture' = str
    else:
        positives.append(False)
print(positives)
```

```python
gender_codes=[]
for a in range(10):
    gender_codes.append(random.randrange(0,2))
print(gender_codes)

gender=[]
for b in gender_codes:
    if b == 0:
        gender.append('Male')
    elif b == 1:
        gender.append('Female')
print(gender)
```

# list comprehension(리스트 내포)

* 빈 리스트에 1부터 10까지 정수를 오름차순 순서로 저장.

```python
numbers=[]
for x in range(1, 11):
    numbers.append(x)
print(numbers)
```

```python
numbers2 = [x for x in range(1,11)] #list comprehension
print(numbers2)
```

```python
squares=[]
for x in range(1,11):
    squares.append(x ** 2)
print(squares)
```

```python
squares2= [x **2 for x in range(1,11)]
print(squares2)
```

* numbers 이름의 리스트에[0,100]범위의 난수 10개를 저장

```python
numbers= [random.randrange(100) for _ in range(10)]
print(numbers)
```

```python
evens = [x for x in numbers if x % 2 ==0]
print(evens)
```

```python
fruits = ['apple', 'banana','cherry','durian','watermellon']
fruit_lengths = [len(x) for x in fruits]
print(fruit_lengths)
```

* [0,100] 범위의 정수 난수 10개를 갖는 리스트를 만듦.
* 리스트의 원소가 짝수이면 'even', 홀수이면 'odd'를 갖는 리스트를 만듦.

```python
# if - else expression(표현식)
n =1
is_even = 'even' if n % 2 ==0 else 'odd'
print(is_even)
```

```python
numbers = [random.randrange(100) for x in range(10)]
print(numbers)
```

```python
result = ['even' if x % 2 == 0 else 'odd' 
          for x in numbers]
print(result)
```

```python
rand = [random.randrange(0,2) for x in range(10)]
print(rand)

result1 =['Male' if x ==0 else 'Female' 
          for x in rand]
print(result1)
```

# 다(2차원) 리스트

```python
# 2차원 리스트: 1차원 리스트들을 원소로 갖는 리스트.
matrix = [[1,2,3],
          [4,5,6],
          [7,8,9]]
print(matrix)
```

## 2차원 리스트의 indexing

```python
matrix[0]
```

```python
matrix[0][0]
```

## 2차원 리스트에서 for 반복문

```python
for row in matrix:
    print(row)
```

```python
for row in matrix:
    for x in row:
        print(x, end='\t') #한 줄에 출력
    print() #줄바꿈
```

```python
m1 = [[1,2,3],
      [4,5,6]]
m2 = [[1,1,1],
      [2,2,2]]
print(m1)
print(m2)
```

```python
adds=[] #리스트들을 저장하기 위한 빈 리스트
for a,b in zip(m1,m2):
    plus=[] #1차원 리스트 a,b에서 같은 인덱스에 있는 원소들끼리 더한 값을 저장할 리스트
    for x,y in zip (a,b):
        plus.append(x+y)
    adds.append(plus)
print(adds)
```

```python
adds=[]
for a,b in zip(m1,m2):
    plus=[x+y for x, y in zip(a,b)]
    adds.append(plus)
print(adds)
```

```python
adds=[[x + y for x, y in zip (a,b)]
     for a,b in zip(m1,m2)]
print(adds)
```
