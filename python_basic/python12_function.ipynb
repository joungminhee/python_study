# 함수(Function)

* 함수(**funcion**): 기능을 수행하는 코드 블록.
* 인수(**argument**): 함수가 기능을 수행하기 위해서 값이 필요한 경우, 함수를 호출할 때 함수에게 전달하는 값(들).

* 매개변수(**parameter**): 함수를 호출 할 때 전달하는 값들(argument)을 저장하기 위해서 함수를 정의할때 선언하는 변수.
* 반환값 (**return value**) : 함수가 기능을 수행한 후에 그 결과로 함수를 호출한 곳으로 반환해 주는 값.
 * 반환값이 있는 함수도 있고, 반환값이 없는 함수도 있음.

````python
result = len([1,2,3,4,5])
print(result)

#함수 :len
# argument: [1,2,3,4,5] -> parameter(obj)에 저장되는 값
# 함수 호출 : len([1,2,3,4,5])
# 반환값(return value) : 5
````

````python
result = print('안녕하세요.')
print(result)

# 함수:print
# argument:'안녕하세요.'
# 함수 호출 : print('안녕하세요.')
# 반환값: None- print 함수는 기능 수행 중에 아무런 반환을 하지 않는다
````

## 2개 이상의 값을 반환하는 함수

````python
def plus_and_minus(x,y):
    """2개 숫자의 x+y 값과 x-y 값을 반환하는 함수."""
    return x+y,x-y
````

````python
result= plus_and_minus(1,2)
print(result) #리턴되는 값은 2개원소를 갖는 tuple
````

````python
a,b = plus_and_minus(1,2) #tuple decomposition(분해) 이용 : a,b = (3,-1)
print(a,b)
````

## 값을 반환하지 않는 함수

````python
def repeat_message(message,n):
    """메시지를 n번 반복해서 출력하는 함수.
    
    @param message:str.
    @param n:int.0 또는 양의 정수.
    @return: None
    """
    for _ in range(n):
        print(message)

    return None 
    #반환값이 없는 함수인 경우, 보통 return None 문장은 생략함.
````

````python
repeat_message('안녕하세요',3)
````

````python
import random
````

````python
def make_list(start,end,n):
    numbers=[] #n개의 난수를 저장할 빈 리스트
    for _ in range(n): #n번 반복하면서
        #(start,end) 범위의 정수 난수를 리스트에 추가
        numbers.append(random.randrange(start,end))

    return numbers 
````

````python
make_list(0,100,10)
````

````python
make_list(0,10,5)
````

# Exercise

* 함수 이름: word_count
* 기능: 문자열들을 원소로 갖는 리스트를 전달받아서, 리스트에 문자열이 몇 번 등장하는 지를 key와 value로 갖는 dictionary를 리턴.
* `result = word_count(['list', 'dict', 'list', 'tuple', 'tuple', 'tuple'])
print(result)` >>> `{'list': 2, 'dict': 1, 'tuple': 3}`

````python
#dictionary comprehension 숙제
word_count=['list', 'dict', 'list', 'tuple', 'tuple', 'tuple']

word={}
for x in word_count:
    word[x]=word_count.count(x)

print(word)
````

````python
#dictionary comprehension 숙제
word2={x : word_count.count(x) for x in word_count}

print(word2)
````

````python
#내가 하다가 못한거
def w_c(x):
    """개수 세어주는 함수."""
    return x.count
````

````python
#선생님이 한거
def word_count(array):  #파라미터 호출 
    counts = {} #문자열을 키로 하고 문자열의 등장 횟수를 값으로 하는 dict를 만들기 위해서
    for x in array: #argument로 전달된 리스트에서 문자열을 하나씩 꺼내면서 반복
        if x in counts: #리스트의 문자열이 dict에 있으면(dict의 키로 존재하면)
            counts[x] += 1 # 문자열의 등장 횟수를 1 증가
        else: #리스트의 문자열이 dict에 없으면(dict의 키로 존재하지 않으면)
            counts[x]=1 #문자열이 처음 등장한 경우이므로, 등장 횟수는 1
    return counts
````

````python
word_count(['a','b','c','a','a'])
````
