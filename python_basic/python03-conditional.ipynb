# 조건문 (Conditional Statement)

## else 없는 조건문

```
if 조건식 :
   조건을 만족할 때 실행할 코드 블록(문장들).
```

* `if 조건식` 뒤에는 반드시 콜론 (:) 을 써야 함. :은 문장이 아직 끝나지않았다
* if 블록은 반드시 들여쓰기(indentation)를 해야함.
* 들여쓰기의 크기는 일정해야함.

````python
age = -16
if age >= 0:
  print(f'당신의 나이는 {age}살이군요.') #(1) if(내부)블록
print('end...') #(2) if 블록이 끝났을 때 실행할 문장.
````

## else 있는 조건문

```
if 조건식 :
    조건을 만족할 때 실행할 코드 블록(문장들)
else:
    조건을 만족하지 못할 때 실행할 코드 블록(문장들)
```

````python
age = -16
if age >= 0 :
    print('정상적인 나이 입력값') #(1)
else :
    print('잘못된 나이 입력값') #(2)
print('end...')
````

# Chained Conditionals

```
if 조건식 1:
    (조건식1)을 만족할 때 실행할 코드 블록
elif 조건식2:
    (조건식2)을 만족할 때 실행할 코드 블록
...
else:
    위의 모든 조건식들을 만족하지 못할 때 실행할 코드 블록
```

* 모든 블록들은 들여쓰기를 해야 함.
* elif 블록은 1개 이상이 있을 수 있음.
* else 블록은 생략될 수 있음.

````python
# 정수 3개를 입력받고 변수에 저장

score1 = int(input('정수 입력>>'))
score2 = int(input('정수 입력>>'))
score3 = int(input('정수 입력>>'))

#정수 3개의 평균을 계산하고 출력
mean = (score1+score2+score3)/3
print ('평균', mean)

# A,B,C,F 결정
if mean >= 90 :
    print ('A')
elif mean >= 80:
    print ('B')
elif mean >=70:
    print ('C')
else : 
    print ('F')
````

* 숫자(실수) 1개를 입력 받고 변수에 저장.
* 숫자 양수, 음수, 0인지를 출력.

````python
fl = float(input())

if fl >0 :
    print ('양수')
elif fl ==0:
    print ('0')
elif fl <0 :
    print ('음수')
````

## 중첩 조건문(nested conditionals)

* if 블록, elif 블록, else 블록 안에서 또다른 조건문 (if, if-else, if-elif-else)을 사용하는 것.
* 들여쓰기에 특히 주의! 들여쓰기 방법에 따라서 완전히 다른 실행 코드(로직)이 될 수 있음.
* elif 또는 논리 연산자(and, or)를 사용하는 문장을 중첩조검누으로 만들수 있음.

```
if 조건식1:
    if 조건식2:
        코드 블록 (1) #조건식1과 조건식2가 동시에 만족할 때
    else:
        코드 블록 (2) #조건식1은 만족하는데, 조건식2는 만족하지 못할 때

if 조건식1:
    코드 블록(1)
else:
    if 조건식2:
        코드 블록 (2)
    else:
        코드 블록 (3)
```

````python
number = float(input('숫자를 입력하세요 >>'))

if number >0:
    print ('양수')
else: #number가 0보다 크지 않으면
    if number ==0:
        print('zero')
    else:
        print('음수')
````

````python
x = int(input('정수 입력 >> '))
y = int(input('정수 입력 >> '))
z = int(input('정수 입력 >> '))

mean = (x+y+z) /3
print('mean >> ',mean)

if mean >= 90:
    print ('A')
else:
    if mean >=80:
        print ('B')
    else:
        if mean >=70:
            print ('C')
        else:
            print ('F')
````
