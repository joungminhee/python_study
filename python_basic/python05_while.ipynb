# while loop(반복문)

```
[초기화 문장]
while 조건식: 
    조건식을 만족하는 동안에 반복할 코드 블록
    [조건을 변경할 수 있는 문장] #[] 생략가능
```

* **들여쓰기 중요!**
* `while 조건식` 다음에는 반드시 콜론(:)을 사용
* 조건식을 만족하는 동안에 블록 내부를 계속 반복 실행 

````python
print (1)
print (2)
print (3)
print (4)
print (5)
````

````python
n =1
while n <=5: #변수 n의 값이 5보다 작거나 같은 동안에
    print(n) #변수 n의 값을 출력
    n +=1 # n = n+1, 변수 n의 값을 1씩 증가.
````

````python
n = 5
while n >=1:
    print(n)
    n -=1
````

````python
# 2 4 6 8 10 출력

n = 2 # n은 2부터 시작
while n <=10: # n의 값이 10보다 작거나 같은 동안 반복
    print(n) #n을 출력
    n +=2 # n의 값에 2씩 증가
````

````python
# 9 7 5 3 1 출력

n= 9
while n >=1:
    print(n)
    n -=2
````

````python
n= 9
while n >=1:
    if n % 2 ==1: #n을 2로 나눈 나머지가 1이면(n이 홀수이면)
        print(n) #홀수일때만 n을 출력
    n-=1 # 홀수/짝수 상관없이 -1을 감소
````

## `break`와 `contiune`

* `break` : 반복(iteration)을 종료. break가 포함된 가장 가까운 반복문을 바로 빠져 나옴.
* `continue`: 다음 반복(iteration)을 수행. **continue아래쪽의 문장들은 건너뛰고** 다음 반복을 수행.


````python
n =1
while n <=5:
    if n ==3:
        break
    print(n)
    n +=1
````

````python
n = 0
while n <5:
    n +=1
    if n ==3 :
        continue
    print(n)
````

## 중첩 while 반복문 (nested while loop)

반복문 안에 또 다른 반복문을 사용하는 것.

````python
dan = 2
while dan <10:
    print(f'---{dan}단---')
    #todo : 출력
    n = 1
    while n <10:
        print(f'{dan}x{n} = {dan * n} ')
        n +=1
    dan +=1
````

````python
dan = 2
while dan < 10:
    n = 1
    while n < 10:
        print(f'{dan} x{n}= {dan * n}')
        if dan == n :
            break # 안쪽 while loop만 종료
        n +=1
    print ('-' * 10) 
    dan +=1
````

## 무한 반복문과 break

````python
while True: # infinte loop(무한 반복문)
    name = input('이름을 입력하세요. 끝내려면 Q를 입력하세요.')
    if name == 'Q':
        print('Good bye~~')
        break
    print(f'안녕하세요, {name}님!')
````
