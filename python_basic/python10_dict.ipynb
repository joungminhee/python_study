#dict

* 사전 (dictionary)형식의 데이터 타입
* 키(key)를 기반으로 값(value)을 저장하는 데이터 타입
 * list,tuple: 인덱스(index)를 기반으로 값들을 저장하는 데이터 타입
 * dict 타입에서 key의 역할은 list나 tuple에서의 인덱스와 마찬가지로, dict에 저장되어 있는 유일한 1개의 값을 참조하기 위한 용도.
   * dict에서 key는 중복된 값을 허용하지 않음.

* `l = [1,2,3]`

* `t = (1,2,3)`

* `d = {1,2,3}`

```python
students = {100: '홍길동', 101: '오쌤', 200:'Scott'}
print(students)
print(type(students))
```

## dict의 메서드

```python
students.keys() # dict.keys(): dict가 가지고 있는 키(key)들의 리스트.
```

```python
students.values() # dict.values(): dict가 가지고 있는 값(values)들의 리스트.
```

```python
students.items()
# dict.items(): (key:value) 튜플들의 리스트.
```

```python
students.get(100)
```

```python
students[100] # get() 메서드의 역할을 인덱스 연산자([])로 할 수 있음.
```

## dict와 for 반복문

```python
book = {'title': '파이썬 라이브러리 데이터 분석',
        'author': '웨스 매키니',
        'price': 35_000}
book
```

```python
for k in book:
    print(k)
# for - in 구문에서 dict를 사용하면 key에 대해서 iteration(순회,반복)을 수행함.
```

```python
# for-in 구문으로 dict의 key와 value를 모두 출력
for k in book:
    print(k , ':', book[k])
```

```python
for k,v in book.items():
    print(k,':',v)    
```

## dictionary comprehension

```python
stu_no = [101,201,302] #학번들의 리스트
stu_name = ['홍길동','허균','오쌤'] #학생들 이름들의 리스트

# 학번을 key로 하고 이름을 value로 하는 dict를 만듦.

students = {} # "학번:이름"을 아이템으로 저장할 빈 사전
for k,v in zip(stu_no,stu_name):
    students[k]=v

print(students)
```

```python
# dict_name = {key: value for ... in ...}
students2= {k:v for k,v in zip (stu_no,stu_name)}
students2
```

```python
fruits = ['apple', 'banana', 'cheery','pineapple','watermelon','딸기']

#과일의 이름을 key로 하고, 이름의 길이를 value로 하는 사전

fruit_len={}
for x in fruits:
    fruit_len[x]=len(x) # fruit_len[과일이름] : 이름의 길이

print(fruit_len)
```

```python
fruit_lengths2= {x : len(x) for x in fruits}

print(fruit_lengths2)
```
