프로그램 실행 중에 오류(error)가 발생하는 경우 해결 방법:

* 오류가 발생한 위치와 원인을 파악해서 오류가 발생하지 않도록 코드를 수정
* 오류가 발생하더라도 프로그램이 (비정상적으로) 종료되지 않고 계속 실행될 수 있도록 처리

```
try:
   (1) 실행할 코드 블록
except [에러 이름 [as 변수]] :
   (2) 에러가 발생했을 때 실행 코드 블록
[else:
   (3) try 블록이 에러 없이 정상적으로 모두 실행됐을 때 실행할 코드 블록]
[finally:
   (4) try 블록 실행 중에 에러 발생 여부에 상관없이 항상 실행할 코드 블록]
```

* except에서 에러 이름을 생략하면, 모든 종류의 에러를 처리할 수 있음.
* except에서 에러 이름을 지정하면, 지정한 이름의 에러만 처리할 수 있음.
* 처리해야 할 에러의 종류가 여러개가 있다면 except 구문을 여러개 사용할 수도 있음.
* try 구문의 실행 순서
 * 에러가 발생하지 않은 경우 : (1) > (3) > (4)
 * 에러가 발생한 경우 : (1) > (2) > (4) 

````python
x = float(input('숫자를 입력하세요>>'))
y = float(input('숫자를 입력하세요>>'))
print(f' {x} + {y}={x+y}')
print(f' {x} - {y}={x-y}')
print(f' {x} * {y}={x*y}')
print(f' {x} / {y}={x/y}')
````

````python
try : 
    x = float(input('숫자를 입력하세요>>'))
    y = float(input('숫자를 입력하세요>>'))
    print(f' {x} + {y}={x+y}')
    print(f' {x} - {y}={x-y}')
    print(f' {x} * {y}={x*y}')
    print(f' {x} / {y}={x/y}')
except:
    print('에러 발생...')
````

````python
try : 
    x = float(input('숫자를 입력하세요>>'))
    y = float(input('숫자를 입력하세요>>'))
    print(f' {x} + {y}={x+y}')
    print(f' {x} - {y}={x-y}')
    print(f' {x} * {y}={x*y}')
    print(f' {x} / {y}={x/y}')
except ValueError as e: #valueError말고 exception은 모든 에러들을 처리할때
    print(e)
    print('입력값은 숫자여야 합니다.')
except ZeroDivisionError as e:
    print(e)
    print('0으로는 나눌 수 없습니다.')
````

## try-except 응용

사용자에게 숫자를 입력하라고 알려주고, 숫자로 변환할 수 있는 문자열이면 변환 후에 정상 종료시키고 에러 발생하면 다시 입력하도록 코드작성

````python
while True: #무한 반복문
    try:
        x=float(input('숫자 입력>>>'))
        print(f'x={x}')
        break # while 반복문 종료
    except Exception: #모든 종류의 예외(exception)을 처리
        print('입력한 값은 숫자로 변환할 수 없습니다.')
````
