# For loop

```

for 변수 in iterable-type:
    반복해서 실행할 코드 블록

```

* iterable-type : 순회(반복)이 가능한 데이터 타입.
 * `str`, `range`, `list`, `tuple`, `dict`, ...
* for 반복문은 iterable-type의 객체에서 원소들을 하나씩 꺼내서 변수에 저장하고, 코드 블록을 반복해서 실행
* iterable-type의 객체에서 더이상 꺼낼 원소가 없을 때 반복문이 종료.

````python
for x in '안녕하세요!':
    print(x)
````

````python
# range(start, stop. step): start <= x < stop 범위의 step(양수)만큼씩 증가하는 정수들을 생성
for x in range(1,5,1):
    print(x)
````

````python
# range(start, stop): start <= x < stop 범위의 1씩 증가하는 정수들을 생성.
for x in range(1,5):
    print(x)
````

````python
# range(stop) : 0 <= x < stop 범위의 1씨기 증가하는 정수들을 생성.
for x in range(5):
    print(x)
````

````python
# range(start, stop. step): start >= x > stop 범위의 step(양수)만큼씩 증가하는 정수들을 생성
for x in range(5,0,-1): #range는 시작하는 숫자는 포함하고 끝나는 숫자는 포함하지 않는다
    print(x)
````

## nested for loop(중첩 for 반복문)

````python
for x in range (2,10):
    for y in range (1,10):
        print(f'{x} x {y} = {x*y}')
    print('-'*10)
````

## `break`와 `continue`

* `break`: iteration(반복)을 종료.
* `continue`: (현재 작업을 멈추고) 다음 iteration(반복)을 수행.

````python
for x in range(5):
    if x == 2:
        break
    print(x)
````

````python
for x in range(5):
    if x ==2:
        continue
    print(x)
````

````python
#구구단 2단은 2x2까지, 3단은 3x3까지,...출력
for x in range(2,10):
    for y in range(1,10):
        print(f'{x} x {y} = {x*y}')
        if x == y :
            break
    print('-'*10)
````

````python
for x in range(2,10):
    for y in range(1,x +1):
        print(f'{x} x {y} = {x*y}')
    print('-'*10)
````

# `print()` 함수
* 문자열을 출력할 때 함께 사용되는 특수 기호
 * ' \t ' : 탭(tab)
 * ' \n ' : 줄바꿈 (new line)
* `print()` 함수의 파라미터(parameter)
 * `end` : `print()`함수의 `arguement(인수)`들을 콘솔에 출력한 후 어떤 동작을 할 것인지를 설정. 기본값은 줄바꿈 ('\n')
 * `sep` : separator(구분자). `print()` 함수의 `arguement(인수)`들을 출력할 때 각각의 `arguement(인수)`들을 구분하기 위한 문자를 설정. 기본값은 공백 한 칸 (' ')



````python
print('안녕\thello\tBon Jour')
````

````python
print('안녕\nhello\nBon Jour')
````

````python
print('안녕','hello', 123)
````

````python
print('안녕','hello', 123, sep=',')
````

````python
print('안녕')
print('Hello')
print(123)
````

````python
print('안녕', end='\t')
print('Hello') # end='\n': 기본값
print(123)
````

````python
for x in range(1, 4):
    print(x, 'little', end=' ')
print()
for x in range(4, 7):   
    print(x, 'little', end=' ')
print()
for x in range(7, 10):   
    print(x, 'little', end=' ')
print()
print('10 little indian boys!')
````

````python
for x in range(1, 11):
    print(x, 'little', end=' ')
    if x % 3 == 0:
        print(x, 'indian') #\n이 기본값
    elif x % 10 == 0:
        print('indian boys!')
````

````python
for x in range (1, 101):
    if (x % 10 == 3) or (x % 10 == 6) or (x % 10 == 9):
        # 10으로 나눈 나머지는 1의 자리 숫자:
        # 1의 자리 숫자가(나머지)가 3 또는 6 또는 9이면
        print ('*', end='\t')
        # 숫자 대신 별표 출력
    elif (x // 10 == 3) or (x // 10 == 6)or (x // 10 ==9):
        # 10으로 나눈 몫은 10의 자리 숫자:
        # 10의자리 숫자(몫)가 3 또는 6 또는 9이면
        print('*', end= '\t')
        # 숫자 대신 별표 출력
    else :    
        print(x, end= '\t') 
        # 둘다 아니라면 숫자를 그대로 쓰겠다
    if (x % 10 ==0) :
        print()
````

````python
for n in range(1,101):
    n_str= str(n) # 명시적 타입 변환 -> 문자열 변환
    if ('3' in n_str) or ('6' in n_str) or ('9' in n_str):
        print ('*', end='\t')
    else:
        print(n, end='\t')
    if n % 10 == 0:
        print()
````
