## reshape

```python
arr_1d = np.arange(12) #1차원 배열
print(arr_1d)
```

```python
# 1차원 배열 arr_1d를 (4, 3) shape을 갖는 2D 배열로 변환
arr_2d = arr_1d.reshape((4, 3))
print(arr_2d)
```

```python
# 1차원 배열 arr_1d를 row가 2개인 2D 배열로 변환
# reshape의 argument를 전달할 때, 자동으로 계산할 수 있는 차원의 수는 -1로 대체가능함.
arr_2d = arr_1d.reshape((4, -1))
print(arr_2d)
```

```python
arr_2d = arr_1d.reshape((-1, 4))
print(arr_2d)
```

```python
# 2차원 배열 - > 1차원 배열
result = arr_2d.reshape((12,))
print(result)
```

```python
result = arr_2d.reshape(3*4)
print(result)
```

```python
result = arr_2d.reshape((-1,))
print(result)
```

## `revel` vs `flatten`

```python
print(arr_2d)
```

```python
flattened = arr_2d.flatten()
# 1차원으로 변환된 배열 "복사본"을 리턴.
# -> 원본(2차원)배열과 별개의 배열을 만들어서 리턴.
print(flattened)
```

```python
flattened[0]=123 # 변환된 1차원 배열의 원소를 변경
print(flattened)
print(arr_2d) #원본 2차원 배열의 원소에는 영향을 미치지 않음.
```

# Broadcasting
* np.ndarray의 산술연산(+, -, *, /, //, %, ...)은 같은 위치(인덱스)의 원소들끼리(element-wise) 연산이 수행됨.
* 서로 shape이 다른 배열들 사이에서 산술연산이 가능하도록 shape을 맞춰서 연산하는 방법 -> broadcast

```python
a = np.array([1,2,3])   #(3,) shape의 1d 배열
b = np.array([1,2,3,4]) #(4,) shape의 1d 배열
print(a)
print(b)
# a+b
# ValueError 발생 - broadcast가 불가능
```

## ndarray와 scalar(숫자 한개)의 broadcast 연산

```python
a + 10 
```

```python
a2 = np.arange(6).reshape((2,3))
print(a2)
```

```python
a2+10
```

## broadcast 활용

* 표준화(standardization): 평균이 0이 되고, 표준편차가 1이 되도록 변수들의 스케일을 변환하는 것.
* 정규화(normalization): 최솟값이 0이 되고, 최댓값이 1이 되도록 변수들의 스케일을 변환하는 것.

#표준화

```python
x = np.array([1,2,3,4,5])
```

```python
mu = np.mean(x) #평균
print(mu)
```

```python
sigma = np.std(x) #표준편차
print(sigma)
```

```python
x_scaled = (x - mu) / sigma #표준화
print(x_scaled)
```

```python
print('평균:',np.mean(x_scaled))
print('표준편차:',np.std(x_scaled))
```

# 정규화

```python
x_sclaed = (x- np.min(x)) / (np.max(x) - np.min(x))
print(x_sclaed)
```

### 2차원 배열에서의 표준화, 정규화

```python
# (3,2)
x = np.arange(1,7).reshape((3,2))
print(x)
```

```python
np.mean(x) # axis = None (기본값)
```

```python
np.mean(x, axis=0) # (1,3,5)의 평균, (2,4,6)의 평균
```

```python
np.mean(x, axis=1) # (1,2) 의 평균, (3,4)의 평균, (5,6)의 평균
```

```python
np.mean(x, axis=1, keepdims=True)
```

```python
# 표준화 
mu = np.mean(x, axis=0, keepdims=True) #평균
sigma = np.std(x, axis=0, keepdims=True) #표준편차
x_scaled = (x - mu)/sigma
print(x_scaled)
```

```python
# 정규화
x_max = np.max(x, axis=0, keepdims=True) #최댓값
x_min = np.min(x, axis=0, keepdims=True) #최솟값
x_scaled = (x - x_min)/ (x_max - x_min)
print(x_scaled)
```

* seaborn 패키지의 iris 예제 데이터 셋을 로드
* iris 데이터프레임에서 species 변수를 제거
* iris 각 변수들을 표준화(standardization)
* iris 각 변수들을 정규화(normalization)

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
x = iris.drop(columns='species').values
x
```

```python
x.index #row label
```

```python
x.columns #column label
```

```python
x= x.values #값들로만 이루어진 2차원 배열(2d ndarray)
```

```python
x.shape
```

```python
# 표준화
mean = np.mean(x,axis=0,keepdims=True) #평균
std = np.std(x, axis=0, keepdims=True) #표준편차
print(mean)
print(std)
```

```python
x_stand = (x-mean)/std # 각 변수들끼리 표준화
print(x_stand[:5, :])
```

```python
np.mean(x_stand, axis=0)
```

```python
np.std(x_stand, axis=0)
```

```python
# 정규화
x_max = np.max(x, axis=0, keepdims=True) # 각 변수들에서 최댓값
#keepdims=True 차원을 유지하겠다.
x_min = np.min(x, axis=0, keepdims=True) # 각 변수들에서 최솟값
x_nor = (x-x_min)/ (x_max-x_min) # 정규화(min-max scaling)
x_nor[:5,:] #:5 행 , : 컬럼
# 최솟값 0 ~ 최댓값 1 사이에 균등하게 분포하는것이 정규화
```

```python
np.max(x_nor, axis=0)
```

```python
np.min(x_nor, axis=0)
```

## 2차원 배열에서의 행렬 곱

`(i,j) @ (j,k) = (i,k)`

```python
a = np.arange(1,7).reshape((2,3))
print(a)
```

```python
b = np.ones(shape=(3,2))
print(b)
```

```python
a @ b # (2,3) @ (3,2) = (2,2)
```

```python
b @ a # (3,2) @ (2,3) = (3,3) # 행렬 곱이 되려면 2,2이 같아야하고 최종으로는 3,3이 나옴
```

## 2차원 배열과 1차원 배열의 행렬 곱

`(i,j) @ (j,)= (i,) # 2D @ 1D = 1D`

`(i,j) @ (j, 1) = (i,1) # 2D @ @2D = 2D`

```python
a # (2,3) shape
```

```python
b = np.array([1,2,1]) # (3, ) shape
```

```python
b.shape
```

```python
a@b # (2,3) @ (3,) = (2,)
```

```python
c = np.array([[1],
             [2],
             [1]]) # (3,1)
```

```python
a @ c # (2,3) @ (3,1) = (2, 1)
```

```python
a @ b [:, np.newaxis] #  == a @ c
```
