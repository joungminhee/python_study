# JSON

* JSON( JavaScript Object Notation): 자바스크립트 객체 표현법.
* `{key1: value1, key2: value2, ...}` 형식의 **문자**열
* Python의 `dict` 클래스 객체의 문자열 표현과 비슷
* `json` 모듈 : JSON 형식 문자열(str) <--> `dict` 타입 객체 (object)
 * `json.dumps(obj)` : python 객체 obj를 json 형식의 문자열로 변환(serialization, 직렬화) 해서 리턴.
 * `json.loads(s)`: json 형식의 문자열(s)를 python 객체 obj로 변환(de-serialization, 역직렬화) 해서 리턴.
 * `json.dump(obj,fp)` : python 객체 obj를 문자열로 변환해서 파일 fp에 씀(write).
 * `json.load(fp)` : 파일 fp에서 json 형식의 문자열을 읽어서 파이썬 객체 obj로 변환하고 그 객체를 리턴.

# Serialization

```python
# 직렬화 : Python 객체 ----> Json 문자열
person = {
    'name': '홍길동',
    'age': 16,
    'phones': ['02-0000-0000', '010-1111-1111']
}
```

```python
type(person) #> dict 타입(객체)
```

```python
person_json =json.dumps(person)
person_json
```

```python
type(person_json) #> str 타입 (문자열)
```

#De-serialization

```python
# 역직렬화: json 형식의 문자열 ---> python 객체(object)
result = json.loads(person_json)
```

```python
type(result)
```

#JSON 파일 쓰기(write)/ 읽기(read)

* picke: binary(이진) 파일 읽기/쓰기
* json: text(텍스트) 파일 읽기/쓰기

```python
with open('gildong.json',mode='wt') as f: #파일을 '텍스트 쓰기' 모드로 오픈.
    # mode= 'rt'기본값, mode='w'와 mode='wt'는 같은 의미.
    f.write(json.dumps(person)) #person 객체를 json 문자열로 변환 후 파일에 씀 (write).
    #with-as 구문 종료 후에는 f.close()자동으로 호출됨.
```

```python
with open('gildong.json') as f: # 파일을 '텍스트 읽기(rt)' 모드로 오픈
    line=f.readline() # 파일의 첫번째 줄(line)을 읽음.
    print(line) # 화면에 출력
    obj = json.loads(line) #파일에서 읽은 문자열을 객체로 변환
    print(obj)
```

```python
file_name ='movies.txt'
with open(file_name, mode='w')as f:
    # dict 객체 생성
    movie1 = {'mid': 1, 'title':'브로커','genre':'drama'}
    movie2 = {'mid': 2, 'title':'범죄도시2','genre':'action'}
    movie3 = {'mid': 3, 'title':'짱구는 못말려', 'genre':'animation'}
    # dict ---> JSON formatted string
    m1_s = json.dumps(movie1)
    m2_s = json.dumps(movie2)
    m3_s = json.dumps(movie3)
    # 3개의 문자열을 한 줄씩 파일에 write
    f.write(m1_s + '\n')
    f.write(m2_s + '\n')
    f.write(m3_s + '\n')
```

```python
# movies.txt 파일에서 한 줄씩 읽으면서, 읽은 문자열을 dict 객체로 변환하고 리스트에 추가.
movies = []
with open(file_name) as f: # 파일에서 한줄씩 읽으면서 반복
    for line in f:
        obj = json.loads(line.strip())  # json formatted string ---> dict
        # strip() 문자열 양 끝의 공백(\s,\t,\n)을 모두 제거
        movies.append(obj)

movies #> dict 객체를 아이템으로 갖는 리스트
```

```python
# list comprehension
with open(file_name) as f:
    movies = [json.loads(line.strip())for line in f]
movies
```

```python
df= pd.DataFrame(movies)
df
```

## 사용자 접속 환경(OS) 분석
* 데이터프레임의 a 컬럼의 문자열을 분석하면 OS 정보를 일부 알 수 있음. Windows / Non-Windows로 분류
* NA 처리 방법:

 * NA를 다른 값으로 대체
 * 평균값 대체, 중앙값 대체, 특정값(예: 0) 대체
 * 이전 행의 값으로 대체, 다음 행의 값으로 대체
 * NA가 포함된 행(row)을 삭제
 * NA가 많지 않은 경우에(행들을 삭제해도 경향이 많이 바뀌지 않을 때)

```python
df_clean.info()
```

```python
# a 컬럼이 NA(NAN,Null)인 행들을 삭제

df_clean = df_clean[df_clean['a'].notnull()]
```

* `pd.Series.notnull()` : 컬럼의 값이 Null이 아니면 True를 리턴.
* `pd/Series.isnull()` : 컬럼의 값이 Null이면 True를 리턴.

```python
df_clean.info()
```

```python
# 'os' 파생 변수(컬럼)추가
# a 컬럼에 'Windows' 문자열이 포함되어 있으면 Windows, 그렇지 않으면 other를 저장.
# np.where(조건식, 참일때 선택할 값, 거짓일때 선택할 값)
df_clean['os'] = np.where(df_clean['a'].str.lower().str.contains('windows'),
                          'Windows', 'Other')
```

```python
df_clean.os.value_counts()
```

```python
sns.countplot(df_clean.os)
plt.show()
```

## 사용자 시간대(timezone)별, 운영체제(OS)별 빈도수, 상위 10개 시각화

```python
tz_os_counts = df_clean.groupby(['tz','os']).size()
```

```python
tz_os_counts = tz_os_counts.reset_index(level=-1) #가장 마지막 레벨의 인덱스만 컬럼으로 변환
tz_os_counts
```

```python
indexer = df_clean['tz'].value_counts().head(10).index
indexer
```

```python
result = tz_os_counts.loc[indexer]
```

```python
result = result.reset_index()
```

```python
result.columns = ['tz','os','count'] # 데이터 프레임의 컬럼이름
result.head()
```

```python
sns.barplot(x='count',y='tz', hue='os', data=result)
plt.show()
```

```python
# unstack: Series에서 인덱스의 값을 컬럼 이름으로 변환. long ---> wide
df_clean.groupby(['tz', 'os']).size().unstack()
# df_clean.pivot_table(index='tz',columns='os',aggfunc='size') 위 코드와 같음
```

```python
# unstack 과정에서 NA들이 생기게 됨 ---> NA를 0으로 대체
df_clean.groupby(['tz', 'os']).size().unstack().fillna(0)
```

```python
result = df_clean.pivot_table(index='tz',columns='os',aggfunc='size', fill_value=0)
result
#> fill_value: 피벗 테이블을 만들 때 NA가 생기는 경우, NA 대신 채워주는 값
```

```python
result['total'] = result['Other']+ result['Windows']
result.head()
```

```python
df = result.sort_values(by='total', ascending=False).head(10)
df
```

```python
df = df.drop(columns='total')
df.head()
```
